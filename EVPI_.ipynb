```python
import pandas as pd

#   
p_good = 0.7
p_poor = 0.3

# Payoff  
payoff_table = {
    'Good Foreign Conditions': [800000, 1300000, 320000],
    'Poor Foreign Conditions': [500000, -150000, 320000]
}

decisions = ['Expand', 'Maintain status quo', 'Sell now']

# Payoff  
payoff_df = pd.DataFrame(payoff_table, index=decisions)

# 1. Perfect information   
# Good condition Maintain status quo , Poor condition Expand 
EV_perfect = (1300000 * p_good) + (500000 * p_poor)

# 2.    ( )
#    (   )
EV_without_perfect = (1300000 * p_good) + (-150000 * p_poor)  # Maintain status quo

# 3. EVPI 
EVPI = EV_perfect - EV_without_perfect

#  
print(f"Expected Value with Perfect Information (EV_perfect): ${EV_perfect:,.2f}")
print(f"Expected Value without Perfect Information (EV_without_perfect): ${EV_without_perfect:,.2f}")
print(f"Expected Value of Perfect Information (EVPI): ${EVPI:,.2f}")
```
